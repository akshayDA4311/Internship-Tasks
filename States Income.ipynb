```python
#Merging the data
merged = map_df.set_index('st_nm').join(df.set_index('State/ UT'))
```

```python
merged['Per_Capita_Income'] = merged['Per_Capita_Income'].replace(np.nan, 0)
#merged.head(29)
```

```python
#Create figure and axes for Matplotlib and set the title
fig, ax = plt.subplots(1, figsize=(8, 8))
ax.axis('off')
ax.set_title('State wise per capita Income', fontdict={'fontsize': '20', 'fontweight' : '10'})
# Plot the figure
merged.plot(column='Per_Capita_Income',cmap='YlOrRd', linewidth=0.9,ax=ax, edgecolor='0.1',legend = True,legend_kwds = {'label':"Income"})
```
